```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules,fpgrowth
```

```python
t_encoder = TransactionEncoder()
```

```python
tras_data = t_encoder.fit(unique_data["products"]).transform(unique_data["products"])
```

```python
tras_data
```

```python
t_encoder.columns_
```

```python
encode_data = pd.DataFrame(tras_data,columns=t_encoder.columns_)
```

```python
encode_data
```

```python
apri = apriori(encode_data,min_support=0.159, use_colnames=True)
```

```python
association_rules(apri,metric="confidence")
```

```python
# FB -Growth

fpg = fpgrowth(encode_data, min_support=0.059, use_colnames=True)
association_rules(fpg,metric='confidence')
```

```python
from itertools import chain
```

```python
def show_recom(cust_id,number_of_rec):
    products_overall = df["productsID"].unique()
    procust_bought =df[df["CustomerId"]==cust_id]["productsID"].values
    prod_used_for_predict = [i for i in products_overall if i not in procust_bought]
    predict_recom = [(prod, model.predict(cust_id,prod).est) for prod in prod_used_for_predict]

    output_recom = [df[df["productsID"]== i_d]["products"].values[0] for i_d, a in predict_recom[:number_of_rec]]
    out_list = list(chain.from_iterable(output_recom))
    out_list = list(set(out_list))
    return out_list
```
